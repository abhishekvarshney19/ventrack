```python
import time
import pandas as pd

start = time.time()

Vendor_Sales_Summary = pd.read_sql_query("""
WITH 
freight_summary AS (
    SELECT 
        VendorNumber, 
        SUM(Freight) AS FreightCost
    FROM vendor_invoice
    GROUP BY VendorNumber
),
purchase_summary AS (
    SELECT
        p.VendorNumber,
        p.VendorName,
        p.Brand,
        p.Description,                                 
        AVG(p.PurchasePrice) AS AvgPurchasePrice,
        SUM(p.Quantity) AS TotalPurchaseQuantity,
        SUM(p.Dollars) AS TotalPurchaseDollars,
        AVG(pp.Price) AS AvgActualPrice,
        SUM(pp.Volume) AS TotalVolume
    FROM purchases p
    JOIN purchase_prices pp
        ON p.Brand = pp.Brand
    WHERE p.PurchasePrice > 0
    GROUP BY
        p.VendorNumber, p.VendorName, p.Brand, p.Description
),
sales_summary AS (
    SELECT
        VendorNo,
        Brand,
        SUM(SalesDollars) AS TotalSalesDollars,
        SUM(SalesPrice) AS TotalSalesPrice,
        SUM(ExciseTax) AS TotalExciseTax,
        SUM(SalesQuantity) AS TotalSellsQuantity
    FROM Sales
    GROUP BY VendorNo, Brand
),
Vendor_Sales_Summary AS (
    SELECT
        ps.VendorNumber,
        ps.VendorName,
        ps.Brand,
        ps.AvgPurchasePrice,
        ps.TotalPurchaseQuantity,
        ps.TotalPurchaseDollars,
        ps.AvgActualPrice,
        ps.TotalVolume,
        ps.Description,
        ss.TotalSalesDollars,
        ss.TotalSalesPrice,
        ss.TotalExciseTax,
        ss.TotalSellsQuantity,
        fs.FreightCost,
        (IFNULL(ss.TotalSalesDollars, 0) 
         - IFNULL(ps.TotalPurchaseDollars, 0)  
         - IFNULL(fs.FreightCost, 0)) AS EstimatedProfit
    FROM purchase_summary ps
    LEFT JOIN sales_summary ss
        ON ps.VendorNumber = ss.VendorNo AND ps.Brand = ss.Brand
    LEFT JOIN freight_summary fs
        ON ps.VendorNumber = fs.VendorNumber
)
--  Final SELECT outside the CTE
SELECT *
FROM Vendor_Sales_Summary
ORDER BY TotalPurchaseDollars DESC;
""",conn)

end = time.time()
execution_time = (end - start) / 60  # Convert seconds to minutes

print(f" Query executed successfully in {execution_time:.2f} minutes")
display(Vendor_Sales_Summary.head())
```

```python
Vendor_Sales_Summary.fillna(0,inplace= True)
```

```python
Vendor_Sales_Summary['VendorName']= Vendor_Sales_Summary['VendorName'].str.strip()
```

```python
Vendor_Sales_Summary['grossprofit'] = Vendor_Sales_Summary['TotalSalesDollars']-Vendor_Sales_Summary['TotalPurchaseDollars']
```

```python
Vendor_Sales_Summary['grossprofit'].min()
```

```python
Vendor_Sales_Summary['ProfitMargin']= (Vendor_Sales_Summary['grossprofit']/Vendor_Sales_Summary['TotalSalesDollars'])*100
```

```python
Vendor_Sales_Summary['SalesToPurchaseRatio'] = Vendor_Sales_Summary['TotalSalesDollars']/Vendor_Sales_Summary['TotalPurchaseDollars']
```

```python
Vendor_Sales_Summary['stockturnover'] = Vendor_Sales_Summary['TotalSellsQuantity']/Vendor_Sales_Summary['TotalPurchaseQuantity']
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS Vendor_Sales_Summary (
    VendorNumber INTEGER,
    VendorName VARCHAR(100),
    Brand VARCHAR(100),

    AvgPurchasePrice DECIMAL(10,2),
    TotalPurchaseQuantity INTEGER,
    TotalPurchaseDollars DECIMAL(15,2),

    AvgActualPrice DECIMAL(10,2),
    TotalVolume DECIMAL(15,2),

    Description VARCHAR(200),

    TotalSalesDollars DECIMAL(15,2),
    TotalSalesPrice DECIMAL(15,2),
    TotalExciseTax DECIMAL(15,2),
    TotalSellsQuantity INTEGER,

    FreightCost DECIMAL(15,2),
    EstimatedProfit DECIMAL(15,2),
    grossprofit DECIMAL(15,2),
    ProfitMargin DECIMAL(10,4),
    SalesToPurchaseRatio DECIMAL(10,4),
    stockturnover DECIMAL(10,4),

    PRIMARY KEY (VendorNumber, Brand)
);
""")

conn.commit()
print(" Vendor_Sales_Summary table created successfully.")
```
